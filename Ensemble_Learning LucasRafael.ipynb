# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `MNIST handwritten digits` (arquivos de treino e teste)
Realize a separação do conjunto de treino como treino e validação
Utilize `train_test_split` com controle de aleatoriedade (seed)
Retorne: `X_train`, `X_val`, `y_train`, `y_val`

Depois responda:
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def load_data(seed=42):

    train_df = pd.read_csv("/mnt/data/mnist_train.csv")
    test_df = pd.read_csv("/mnt/data/mnist_test.csv")

    X = train_df.drop("label", axis=1).values
    y = train_df["label"].values

    X_train, X_val, y_train, y_val = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )

    return X_train, X_val, y_train, y_val
```

**Para Random Forest e árvores em geral, não é necessário normalizar os dados.**

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def train_random_forest(X_train, y_train, seed=42):
    model = RandomForestClassifier(
        n_estimators=100,
        random_state=seed,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    return model
```

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

```python
from sklearn.metrics import accuracy_score
```

```python
def evaluate(model, X_test, y_test):

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    return acc
```

**Essa função usa o modelo treinado para prever (predict), compara com os valores reais, retorna a acurácia (sem imprimir, mantendo o padrão das anteriores)**.

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

```python
def run_pipeline(model_type="rf", seed=42):

    X_train, X_val, y_train, y_val = load_data(seed)

    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    else:
        raise ValueError("Modelo não suportado")

    acc = evaluate(model, X_val, y_val)

    return acc
```

**1 - Não existe uma profundidade exata universal, isso depende dos dados mas, em geral o max_depth aumenta conforme o modelo começa a memorizar o treino**

**2 - Porque o max_depth=None permite crescimento ilimitado e a árvore continua dividindo até que cada folha fique quase pura ou cada amostra fique isolada**

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest

## Apresente:
- Acurácia, Precisão, Recall e F1-Score para o modelo

**Solução**:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate_metrics(model, X_test, y_test):
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    recall = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)

    return acc, precision, recall, f1
```

```python
def load_data(seed=42):
    import pandas as pd
    from sklearn.model_selection import train_test_split

    train_df = pd.read_csv("mnist_train.csv")
    test_df = pd.read_csv("mnist_test.csv")

    X = train_df.drop("label", axis=1).values
    y = train_df["label"].values

    X_train, X_val, y_train, y_val = train_test_split(
        X, y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )

    return X_train, X_val, y_train, y_val
```

```python
X_train, X_val, y_train, y_val = load_data(seed=42)

model = train_random_forest(X_train, y_train, seed=42)

acc, precision, recall, f1 = evaluate_metrics(model, X_val, y_val)

print(f"Acurácia: {acc:.4f}")
print(f"Precisão: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

# Questão 6

Execute o pipeline utilizando diferentes seeds (ex: 42 e 7).

## Analise:
- Os resultados mudaram?

## Responda:
- O experimento é reprodutível? Justifique.

**Solução**:

```python
X_train, X_val, y_train, y_val = load_data(seed=42)
model_42 = train_random_forest(X_train, y_train, seed=42)
acc_42, prec_42, rec_42, f1_42 = evaluate_metrics(model_42, X_val, y_val)

X_train, X_val, y_train, y_val = load_data(seed=7)
model_7 = train_random_forest(X_train, y_train, seed=7)
acc_7, prec_7, rec_7, f1_7 = evaluate_metrics(model_7, X_val, y_val)

print("Seed 42:")
print(f"Acurácia: {acc_42:.4f}, Precisão: {prec_42:.4f}, Recall: {rec_42:.4f}, F1: {f1_42:.4f}")

print("\nSeed 7:")
print(f"Acurácia: {acc_7:.4f}, Precisão: {prec_7:.4f}, Recall: {rec_7:.4f}, F1: {f1_7:.4f}")
```

**Sim, o experimento é reprodutível desde que a seed seja fixada porque, usamos random_state no train_test_split e usamos random_state no modelo (Random Forest) fazendo com que ao executar novamente com a mesma seed, os resultados serão iguais.**

# Questão 7

Para pelo menos um dos modelos:

- Compare a acurácia em treino e teste

## Responda:
- Existe overfitting?
- Qual modelo tende a sofrer mais com isso?

```python
X_train, X_val, y_train, y_val = load_data(seed=42)

model = train_random_forest(X_train, y_train, seed=42)

train_acc, train_prec, train_rec, train_f1 = evaluate_metrics(model, X_train, y_train)

val_acc, val_prec, val_rec, val_f1 = evaluate_metrics(model, X_val, y_val)

print("Treino:")
print(f"Acurácia: {train_acc:.4f}, Precisão: {train_prec:.4f}, Recall: {train_rec:.4f}, F1: {train_f1:.4f}")

print("\nValidação:")
print(f"Acurácia: {val_acc:.4f}, Precisão: {val_prec:.4f}, Recall: {val_rec:.4f}, F1: {val_f1:.4f}")
```

**1 - Sim, existe overfitting quando o desempenho no conjunto de treino é significativamente melhor que no de validação.**

**2 - Modelos baseados em árvores, especialmente árvores de decisão individuais, tendem a sofrer mais com overfitting.**

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`

## Analise:
- O desempenho muda significativamente?

```python
seeds = 42
n_estimators_list = [10, 50, 100, 200]

results = []

for n in n_estimators_list:
    X_train, X_val, y_train, y_val = load_data(seed=seeds)

    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(
        n_estimators=n,
        random_state=seeds,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    acc, prec, rec, f1 = evaluate_metrics(model, X_val, y_val)

    results.append((n, acc, prec, rec, f1))

for r in results:
    print(f"n_estimators={r[0]} -> Acurácia: {r[1]:.4f}, Precisão: {r[2]:.4f}, Recall: {r[3]:.4f}, F1: {r[4]:.4f}")
```

**Não, o desempenho não muda significativamente após certo ponto. Observa-se que o aumento de n_estimators melhora o resultado inicialmente, mas depois os ganhos se tornam marginais.**
